**Because of how lenghty the countries are, we will be grouping them by region** 

```python
new=[]
for i in df["Country Name"]:
    if i in a :
        new.append('Africa Eastern and Southern')
    elif i in b:
        new.append('Africa Western and Central')
    elif i in c:
        new.append('Middle East & North Africa')
    elif i in d:
        new.append('Europe & Central Asia')
    elif i in e:
        new.append('East Asia & Pacific')
    elif i in f:
        new.append('Latin America & Caribbean')
    elif i in g:
        new.append('North America')
    elif i in h:
        new.append('South Asia')
    elif i in j:
        new.append('Sub-Saharan Africa')
    elif i in k:
        new.append('Other Regions')
    else:
        new.append('No Region')
df['Region']=new
    
```

```python
cols = list(df.columns)
# Remove 'column_10' from its current position
cols.insert(1, cols.pop(14))

# Reorder the DataFrame
df = df[cols]
```
